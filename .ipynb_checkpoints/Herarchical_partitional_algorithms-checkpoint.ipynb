2. Definir una función de comparación, es decir como se determina que tan parecidos son los objetos

   Para este caso, se pueden considerar las siguientes funciones de distancia:

|Nombre|Función|
|:--:|:--:|
|Minkowski | $$L_q(x_i,x_j)=\sqrt[q]{\sum_{k=1}^N (x_{ik}-x_{jk})^q}$$| 
|**Coseno** (simulitud)   | $$\cos(x_i,x_j)=\frac{\sum_{k=1}^N x_{ik}*x_{jk}}{\sqrt{\sum_{k=1}^N x_{ik}^2} \sqrt{\sum_{k=1}^N x_{jk}^2}}$$|

4. Definir cómo medir la calidad de los grupos encontrados.

Para medir la calidad de los grupos encontrados se utilizan las siguientes medidas: 

    - Para medir qué tan compactos son los clusters (cohesión en cada cluster o medida intra-cluster), se utilizará: 1. La suma de los errores al cuadrado o inercia, y 2. El coeficiente de silueta

# Agrupamiento_Jerarquizado

```python
def anota_dendograma(D, ax):
      bbox_props = dict(fc="white", ec="b",boxstyle='circle') # De éste se calcula el dendograma con los mejores especificaciones.
      ll=list(set(np.array(D['dcoord']).flatten()))
      ll.sort()
      for j, d in zip(D['icoord'], D['dcoord']):
        x = 0.5 * sum(j[1:3])
        y = d[1]
        #plt.plot(x, y, 'sr', markersize=10)
        ax.annotate("$C_{%d}$" %(ll.index(y)+len(X1)-1) , (x, y), xytext=(0, 8),textcoords='offset points',va='top',size=9 
                     ,ha='center', bbox=bbox_props)
```

### Data 1 (Banda Ancha Fija y Móvil)

```python
y1 = data1['País']
list(y1)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from matplotlib.cm import ScalarMappable
from scipy.cluster.hierarchy import linkage, dendrogram

# Assuming you have already defined X1, y1, anota_dendograma functions

# Create a label encoder for country names
label_encoder = LabelEncoder()
y1_numeric = label_encoder.fit_transform(y1)

# Create a 1x4 grid of dendrogram subplots
fig, axes_dendrogram = plt.subplots(1, 4, figsize=(32, 8))
axes_dendrogram[0].set_ylabel('Distancia')

# Create a 1x4 grid of scatter plot subplots
fig, axes_scatter = plt.subplots(1, 4, figsize=(32, 8))

# Enumerate over the 1x4 grid for dendrogram subplots
for i, lk in enumerate(('ward', 'centroid', 'single', 'complete')):
    # Create a new figure explicitly
    plt.figure()
    
    # Create the dendrogram subplot
    Z = linkage(X1, lk)
    Dz = dendrogram(Z, labels=list(y1), leaf_rotation=90)
    plt.title(f'Dendrogram ({lk})')
    anota_dendograma(Dz, plt.gca())

    # # Save the figure with a unique identifier
    # plt.show()

# Enumerate over the 1x4 grid for scatter plot subplots
for i, lk in enumerate(('ward', 'centroid', 'single', 'complete')):
    # Create the scatter plot subplot
    scatter_ax = axes_scatter[i]
    
    # Convert X1 to a NumPy array
    X1_array = X1.to_numpy() if isinstance(X1, pd.DataFrame) else X1
    
    # Use numeric labels for colors
    sc = scatter_ax.scatter(X1_array[:, 0], X1_array[:, 1], c=y1_numeric, cmap='viridis')
    
    # Annotate each point with its label
    for label, x, y in zip(list(y1), X1_array[:, 0], X1_array[:, 1]):
        scatter_ax.text(x, y, label, fontsize=8, ha='right', va='bottom', color='black')

    scatter_ax.set_title(f'Scatter Plot ({lk})')

# Create a colorbar for the scatter plots
cbar = plt.colorbar(ScalarMappable(cmap='viridis'), ax=axes_scatter, ticks=np.unique(y1_numeric))
cbar.set_label('Country')
plt.savefig("dendrograma_rediness_index.jpg")
plt.show()
```

El dendograma muestra la relación jerárquica entre los 18 países incluidos en el ranking de acceso a Internet de la OCDE. En principio, entre mayor sea el número de ramas o mayor la distancia entre los nodos, mayor será la disparidad entre los grupos.  

Claramente, nuestros dendogramas muestran a México y Costa Rica como países muy similares en términos de penetración de los servicios de acceso a Internet. De manera más alejada aunque contigua encontramos a Colombia y Chile.

En un primer nivel R1, los países más similares son: 

- Costa Rica y México

- Eslovaquía y Eslovenía

En un segundo nivel R2, los países más similares y cercanos a la OCDE son los siguientes:

Estados Unidos es, en todos los criterios de jerarquización, un país o elemento disímil de los demás, con los mejores datos de penetración del servicio de Internet. 


# Partiendo_Dendogramas

### Partir el dendograma

El dendrograma muestra la similitud entre la jerarquía de clusters, pero también puede utilizarse para determinar el número de clusters, así como los elementos en cada uno. Esto se puede conseguir mediante un "corte" horizontal a una determinada altura del dendograma, el número de ramas que sobrepasan por encima del promedio será el número de clusters. 

```python
# cortando a una altura de 80 tendríamos dos clusters
Z=linkage(X1,'ward')
Dz=dendrogram(Z, labels=list(y1), leaf_rotation=90)
plt.axhline(80,0,10,linestyle='dashed')
anota_dendograma(Dz,plt)
plt.savefig("partiendo.jpg")
```

```python
# cortando a una altura de 40 tendriamos cinco clusters
Z=linkage(X1,'ward')
Dz=dendrogram(Z, labels=list(y1), leaf_rotation=90)
plt.axhline(40,0,10,linestyle='dashed')
anota_dendograma(Dz,plt)
```

# Algoritmos_Particionales

### K-means
Se usaran dos distancias: 

|Función de comparación| valor para el centroide | Función objetivo &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|
|:---:|:---:|:---:|
|Distancia Euclidiana| media|  $$\min \sum_{k=1}^{K}  \sum_{\forall x_i \in G_k} (x_i-c_k)^2$$|
|Distancia Manhattan | mediana| $$\min \sum_{k=1}^{K} \sum_{\forall x_i \in G_k} \|x_i-c_k\|$$|

Primero se determinará el número de clusters óptimo a través del método de siluete y elbow.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

def calculate_silhouette_score(X, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    return silhouette_avg

max_clusters = 15  
silhouette_scores = [calculate_silhouette_score(X1, n) for n in range(2, max_clusters + 1)]

inertia_values = []
for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X1)
    inertia_values.append(kmeans.inertia_)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
ax1.set_title('Silhouette Score vs. Número de Clusters')
ax1.set_xlabel('Número de Clusters')
ax1.set_ylabel('Silhouette Score')
ax1.grid(True)

ax2.plot(range(2, max_clusters + 1), inertia_values, marker='o')
ax2.set_title('Método Elbow')
ax2.set_xlabel('Número de Clusters')
ax2.set_ylabel('Inercia')
ax2.grid(True)

plt.tight_layout()
plt.savefig('cluster_optimo.jpg')
plt.show()
```

En ambos casos, la metodología propone como el número de clústers óptimos 10.

### Se calculan los cluster con la técnica de K-means

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

K = 10  
kmeans = KMeans(n_clusters=K, random_state=42)
X1['Cluster'] = kmeans.fit_predict(X1)

scatter = plt.scatter(X1['Total_BA_100hab'], X1['Total_BAM_100hab'], c=X1['Cluster'], cmap='viridis', label='Observaciones')
centroids = plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='orange', marker='*', s=100, label='Centroides')

for label, x, y in zip(y1, X1['Total_BA_100hab'], X1['Total_BAM_100hab']):
    plt.text(x, y, label, fontsize=8, ha='right', va='bottom', color='black')

legend_labels = list(y1)
plt.legend(handles=[scatter, centroids], labels=['Observaciones'] + legend_labels, loc='upper right')

plt.title('K-Means para el caso de los países con acceso a Internet 2023')
plt.xlabel('Total_BA_100hab')
plt.ylabel('Total_BAM_100hab')
plt.savefig('kmeans.jpg')
plt.show()
```

### Vamos a hacer lo mismo para el caso del conjunto de datos del network rediness index 2023 (138 países)

```python
max_clusters = 15  
silhouette_scores = [calculate_silhouette_score(X2, n) for n in range(2, max_clusters + 1)]

inertia_values = []
for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X2)
    inertia_values.append(kmeans.inertia_)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
ax1.set_title('Silhouette Score vs. Número de Clusters')
ax1.set_xlabel('Número de Clusters')
ax1.set_ylabel('Silhouette Score')
ax1.grid(True)

ax2.plot(range(2, max_clusters + 1), inertia_values, marker='o')
ax2.set_title('Método Elbow')
ax2.set_xlabel('Número de Clusters')
ax2.set_ylabel('Inercia')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

Los resultados parecen sugerir 12 clusters como óptimos para el caso que nos ocupa.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming X2 is your feature data and y2 is the labels
# X2 should be a DataFrame with features, and y2 should be a Series with labels

# Fit KMeans model
K = 12  # You can adjust the number of clusters based on your requirements
kmeans = KMeans(n_clusters=K, random_state=42)
X2['Cluster'] = kmeans.fit_predict(X2)

# Manually adjust the y-axis positions of centroids
y_adjustment = 10  # You can adjust this value based on your requirements
kmeans.cluster_centers_[:, 1] += y_adjustment

# Visualize the clusters and centroids
fig, ax = plt.subplots(figsize=(12, 8))  # Adjust the figure size as needed
scatter = ax.scatter(X2['Score'], X2['Impact'], c=X2['Cluster'], cmap='viridis', label='Observaciones')
centroids = ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='orange', marker='*', s=100, label='Centroides')

# Annotate each point with its label (display only the first four letters)
for label, x, y in zip(y2, X2['Score'], X2['Impact']):
    short_label = label[:4]  # Display only the first four letters
    ax.text(x, y, short_label, fontsize=8, ha='right', va='bottom', color='black')

# Set legend using shortened labels
legend_labels = [label[:4] for label in y2]
ax.legend(handles=[scatter, centroids], labels=['Observaciones'] + legend_labels, loc='upper right')

ax.set_title('K-Means para el caso de la disponibilidad de redes en el mundo 2023')
ax.set_xlabel('Score')
ax.set_ylabel('Impacto')

plt.show()
```
